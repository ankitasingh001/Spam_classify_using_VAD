```python
import csv
import pandas as pd
import librosa
import soundfile as sf
import numpy as np
```

```python
df_h = pd.read_csv('./data/Hindi_full/Hindi.csv')
df_m = pd.read_csv('./data/Marathi_full/Marathi.csv')
df_ta = pd.read_csv('./data/Tamil_full/Tamil.csv')
df_te = pd.read_csv('./data/Telugu_full/Telugu.csv')
```

```python
data_frames_h = df_h[df_h["AIF"]!=0][df_h["BIN"]!=0]
data_frames_m = df_m[df_m["AIF"]!=0][df_m["BIN"]!=0]
data_frames_ta = df_ta[df_ta["AIF"]!=0][df_ta["BIN"]!=0]
data_frames_te = df_te[df_te["AIF"]!=0][df_te["BIN"]!=0]
```

```python
# !rm -r all_clap_speech
!mkdir -p all_clap_speech/Hindi
!mkdir -p all_clap_speech/Marathi
!mkdir -p all_clap_speech/Tamil
!mkdir -p all_clap_speech/Telugu
```

```python
csv_file = open('metadata_all_clap.csv', 'w', newline='')
writer = csv.writer(csv_file)
writer.writerow(["file_path", "label"])
```

```python
from tqdm import tqdm
```

```python
sample_rate = 16000
time = 5
# all_marathi = []

for lang,data_frames in [('Marathi', data_frames_m),('Hindi',data_frames_h),('Tamil',data_frames_ta),('Telugu',data_frames_te)]:
    for i in tqdm(range(len(data_frames))):
        f_name = data_frames.iloc[i].audioFileName
        f_path = './data/'+lang+'_full/'+lang+'/'+f_name+'.wav'
        y, sr = librosa.load(f_path)
        y_16k = librosa.resample(y, sr, 16000)


        if(len(y_16k)>=sample_rate*time):
            ## if clip more than 5s trim it
            x = y_16k[:sample_rate*time]
        else:
            ## if clip less than 5s pad with 0s
            x = y_16k
            x = np.pad(y_16k,(0,sample_rate*time-len(y_16k)), 'constant', constant_values=(0,0))


        # all_marathi.append(x.reshape(1,-1))
        i = i+(sample_rate*time)
        f_name = f'./all_clap_speech/'+lang+'/'+f_name+'.wav'
        sf.write(f_name, x, sample_rate)
        writer.writerow([f_name, "1"])
```

```python
y, sr = librosa.load('./data/QUT-NOISE/QUT-NOISE/CAR-WINUPB-1.wav')
y_16k = librosa.resample(y, sr, 16000)

sample_rate = 16000
time = 5
all_noise = []

i=0
cnt=0
while((i+(sample_rate*time))<len(y_16k)):
    x = y_16k[i:i+sample_rate*time]
    all_noise.append(x.reshape(1,-1))
    i = i+(sample_rate*time)
    f_name = f'./final_data/noise/CAR-WINUPB-1_trim_{time}s_{cnt}.wav'
    sf.write(f_name, x, sample_rate)
    writer.writerow([f_name, "0"])
    cnt += 1

    if(cnt>150):
        break
```
